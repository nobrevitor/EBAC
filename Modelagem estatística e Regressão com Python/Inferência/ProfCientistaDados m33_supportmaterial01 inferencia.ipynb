## Desgaste de sola de calado

Vamos estudar o nosso teste de hipteses com um problema que consiste no estudo de desgaste de dois tipos de solado de calado.

Um fabricante desenvolveu um solado novo, mais ecolgico, mais barato, e confortvel. Ele quer comparar os desgastes, e acredita que o solado novo  melhor que o tradicional (oferece menor desgaste).

Para isto fez o seguinte desenho amostral: Selecionou 20 crianas de uma escola de baixa renda, todas dispostas a participar do estudo, as quais ficaram felizes por ganhar o par de calados do experimento. Como 10 crianas so de uma sala e as outras 10 so de outra sala, separou as crianas em grupos conforme a sala:

- Grupo A: usa o solado de calado tradicional
- Grupo B: usa o solado de calado novo

Em seguida orientou a todos que utilizassem exclusivamente o calado do experimento, acompanhou as crianas por 6 meses, e ao trmino do perodo fez uma medida do desgaste em micrmetros. Agora ele deseja comparar o desgaste do solado entre os grupos A e B.

```python
np.random.seed(2360873)

solado_a = 0
solado_b = -15

fator_criana = np.random.normal(100, 30, 20)
efeito_p = np.random.normal(0, .5, 20)
solado_novo = np.repeat((1, 0), (10, 10))

p_direito  =  fator_criana + solado_novo*solado_b + efeito_p
p_esquerdo =  fator_criana + solado_novo*solado_b - efeito_p
desenho_1 = pd.DataFrame({'solado':solado_novo, 'p_direito':p_direito, 'p_esquerdo':p_esquerdo})
desenho_1['Grupo']= desenho_1['solado'].map({0:'Grupo A', 1:'Grupo B'})
desenho_1['desgaste_total']= desenho_1['p_direito']+desenho_1['p_esquerdo']
```

### Os dados
Temos ento 20 crianas, as primeiras 10 com o solado novo (Grupo B), as 10 seguintes com o solado tradicional (Grupo A). Temos registrado o desgaste da sola de sapato de cada p, e um desgaste total que consiste na soma de ambos. O grupo tambm est registrado na tabela

### Fazendo o teste

Como j vimos, o teste-t consiste em calcular a estatistica *t* e compar-la com a distribuio *t* padro, com *n-1* graus de liberdade. A estatstica *t* pode ser calculada pela seguinte frmula:

$$ t=\frac{\bar{x_1}-\bar{x_2}}{S_{x_1x_2}\sqrt{\frac{2}{n}}} $$

Sendo que:

$$ S_{x_1x_2} = \sqrt{\frac{S_{x_1}^2 +S_{x_2}^2}{2}} $$

A estatstica *t*, **sob $H_0$** tem a seguinte distribuio:

$$ t \sim t(n-2) $$

O cdigo abaixo replica a frmula do teste-t. 

```python
desenho_1.groupby('Grupo')['desgaste_total'].var().mean()**.5
```

```python
desvpad = desenho_1.groupby('Grupo')['desgaste_total'].var().mean()**.5
Mdia_A = desenho_1.groupby('Grupo')['desgaste_total'].mean()['Grupo A']
Mdia_B = desenho_1.groupby('Grupo')['desgaste_total'].mean()['Grupo B']

diferena = Mdia_A - Mdia_B # Numerador da frmula
erro_padrao = desvpad*((2/10)**.5) # varincia amostral*raiz(2/n)
t = diferena/erro_padrao     # Estatstica do teste
RC = st.t.ppf(.95, 18)        # Regio crtica: sob h0, t est na RC 5% das vezes
p_value=1-st.t.cdf(t, 18)     # p-value: a menor significncia para a qual rejeitamos o teste

print(f'diferena: {diferena:.2f}')
print(f'desvio padro: {desvpad:.2f}')
print(f'erro padro: {erro_padrao:.2f}')
print(f'estatstica do teste: {t:.2f}')
print(f'regio crtica: {RC:.2f} ou mais')
print(f'p-value: {p_value:.5f} ou mais')
```

### Anlise grfica do teste
O cdigo abaixo faz o grfico da distribuio da estatstica de teste, indica a regio crtica, e o valor da estatstica do teste. Conclumos com essa anlise que h diferena estatstica.

```python
fig, ax = plt.subplots(figsize=(10,3))

# Parmetros da densidade sob H0
mu = 0
variance = 1
sigma = variance**.5

# Valores de X e Y para plotar a densidade
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = st.t.pdf(x, 18)

# low = incio da regio crtica do teste
low = st.t.ppf(.95, 18)
high = np.inf

# Grfico da densidade
ax.plot(x, y)
# Esta linha de referncia marca o valor observado da estatstica do teste
ax.vlines(x=[t], 
             ymin=0, ymax=y.max(), 
             colors='r', linestyles='dotted')

# rea 
plt.fill_between(x, y, where=(low < x) & (x < high), alpha=.5)

# Probabilidade de X pertencer  rea
prob_area = st.t.cdf(high, 18) - st.t.cdf(low, 18)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))
```

### Teste-t pelo statsmodels

```python
from statsmodels.stats.weightstats import ttest_ind

x1 = desenho_1.loc[desenho_1['Grupo'] == 'Grupo A', 'desgaste_total']
x2 = desenho_1.loc[desenho_1['Grupo'] == 'Grupo B', 'desgaste_total']

teste_t = ttest_ind(x1, x2, alternative='larger', usevar='pooled', value=0)

teste_t
```

```python
print(f'Estatstica do teste: {teste_t[0]:.2f}')
print(f'             P-value: {teste_t[1]:.4f}')
print(f'  Graus de liberdade: {teste_t[2]:.2f}')
```

## Desenho amostral alternativo

O fabricante ento resolveu chamar um cientista de dados que props um desenho novo: cada p do calado em um material diferente. Assim, todos os efeitos da criana, biolgicos, comportamentais e at de exposio climtica seriam "pareados", e a diferena do desgaste seria dada somente pela diferena do tipo de material.

O fabricante dividiu as crianas em grupos aleatoriamente, mas os grupos indicam agora que solado fica em qual p:

- Grupo A: Solado tradicional no p esquerdo, e novo no p direito. 
- Grupo B: Solado tradicional no p direito, e novo no p esquerdo. 

Ao final do perodo, o desgaste foi registrado. O fabricante computou a diferena entre os desgastes do solado Novo menos o desgaste no Tradicional para cada criana, e se essa diferena for predominantemente negativa, significa que o solado novo  melhor.

O cdigo abaixo simula os dados nessa situao. Vamos ver os resultados em seguida:

```python
np.random.seed(2360873)

solado_a = 0
solado_b = -15

fator_criana = np.random.normal(100, 30, 20)
efeito_p = np.random.normal(0, .5, 20)
solado_novo_dir = np.tile((1, 0), 10)
solado_novo_esq = 1-solado_novo_dir

p_direito  =  abs(fator_criana + solado_novo_dir*solado_b + efeito_p)
p_esquerdo =  abs(fator_criana + solado_novo_esq*solado_b - efeito_p)
desenho_2 = pd.DataFrame({'solado_novo_dir':solado_novo_dir,
                      'solado_novo_esq':solado_novo_esq, 
                      'p_direito':p_direito, 
                      'p_esquerdo':p_esquerdo})
```

### Os dados
Um aspecto chave do desenho 2  que vamos comparar o desgaste em cada indivduo variando apenas o material, que  o que queremos estudar. Para isto, vamos avaliar a diferena do desgaste entre o p direito e o p esquerdo. 

Agora o grupo  uma mera varivel de controle, que indica quem tem o material novo no p direito ou no p esquerdo. No esperamos diferenas por conta disso.

```python
desenho_2['diferena'] = desenho_2['p_direito']*desenho_2['solado_novo_dir'] -\
                      desenho_2['p_esquerdo']*desenho_2['solado_novo_dir'] +\
                     desenho_2['p_esquerdo']*desenho_2['solado_novo_esq']-\
                      desenho_2['p_direito']*desenho_2['solado_novo_esq']

desenho_2['Grupo']= desenho_2['solado_novo_dir'].map({0:'Grupo A', 1:'Grupo B'})
desenho_2.head()
```

### Anlise descritiva

Podemos at avaliar os desgastes originais na anlise descritiva. Mas a anlise que vai responder mesmo a nossa pergunta  a avaliao da diferena entre os desgastes do p direito e do p esquerdo criana a criana. Como elas usam calados com materiais diferentes, a principal fonte de variao  o material do calado.

Repare que o 'importante' neste teste  que os valores sejam diferentes de zero. Por isso 'foramos' o zero a aparecer no grfico para realar a distncia das medidas ao zero.

```python
fig, ax = plt.subplots(1, 2, figsize=(8,4), sharey=True)

sns.boxplot(data=desenho_2, y='diferena', x='Grupo', ax=ax[0])
ax[0].set_title('Boxplot')
ax[0].set_ylim((desenho_2['diferena'].min()*1.1, 0))
ax[0].set_ylabel('Diferena do desgaste: Novo - Tradicional')

sns.pointplot(data=desenho_2, y='diferena', x='Grupo', ax=ax[1])
ax[1].set_title('Mdia e intervalo bootstrap')
ax[1].set_ylabel('')

plt.subplots_adjust(wspace=0)
```

### Teste-t pareado
Nesse caso, o teste recomendado  o teste-t pareado. Ele tem esse nome porque como avaliamos a diferena entre medidas em um mesmo indivduo, **pareamos** todos os demais fatores em cada medida, de modo que a diferena s pode se valer  diferena entre materiais (o tratamento que estamos utilizando).

Os dados que utilizamos no teste so as diferenas para cada indivduo i=1, 2,... n:

$$ D_i = X_{1i}-X_{2i} $$

Em que, no nosso caso, $X_{1i}$  o desgaste medido no p esquerdo da criana i, e $X_{2i}$  o desgaste medido da criana i no p direito, com $i = 1, 2, ..., n$.

Calculamos ento a mdia dessas diferenas e a varincia amostral:

$$ \bar{D} = \sum_{i=1}^n \frac{D_i}{n} $$

$$ S_D = \sum_{i=1}^n \frac {\left(D_i - \bar{D}\right)^2}{n-1} $$

Nesse teste as hipteses so:<br>
- $H_0$: a mdia das diferenas  zero<br>
- $H_a$: a mdia das diferenas  menor que zero

A estatstica do teste :

$$ t = \frac{\bar{X_D}}{S_D\frac{1}{\sqrt{n}}} $$

E, sob $H_0$ a distribuio de $t$ :
$$ t\sim t(n-1) $$

```python
n=desenho_2.shape[0]
D_barra = desenho_2['diferena'].mean()
Sd = desenho_2['diferena'].std()
ep=Sd/(1/n**.5)

t = D_barra/ep

t
```

```python
fig, ax = plt.subplots(figsize=(10,3))

mu = 0
variance = 1
sigma = variance**.5
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y = st.t.pdf(x, 18)
low = -np.inf
high = t

ax.plot(x, y)
ax.vlines(x=[t], 
             ymin=0, ymax=y.max(), 
             colors='r', linestyles='dotted')


plt.fill_between(x, y, where=(low < x) & (x < high), alpha=.5)

prob_area = st.norm.cdf(high, mu, sigma) - st.norm.cdf(low, mu, sigma)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.6%}'.replace('.',','))
```

Este estudo deixa claro que o solado novo possui menor desgaste que o solado antigo, pois teve desgaste menor para todas as crianas. Esse desgaste  considerado relevante, pois  (em valor absoluto) maior que 10 micrmetros.

### Erro tipo II e Poder do teste

At agora, tudo o que fizemos foi sobre o erro tipo I, que  a probabilidade de rejeitar $H_0$ dado que $H_0$  verdadeira.

O erro tipo II  a probabilidade de *no rejeitar* $H_0$ dado que $H_a$  verdadeira. Nesse caso, para podermos fazer os clculos, precisamos de uma hiptese nula *pontual*, sobre a qual podemos fazer contas. Resumindo:

- **Erro tipo I**:  a probabilidade de rejeitar $H_0$ quando ela  verdadeira.
- **Erro tipo II**:  a probabilidade de **no** rejeitar $H_0$ quando ela  **falsa** - e quando especficamente $H_a$  verdadeira.

Denotamos a probabilidade do erro tipo I com a letra grega $\alpha$, da mesma forma, denotamos a probabilidade do erro tipo II com a letra grega $\beta$.

O **poder** do teste  o complementar ao erro tipo II: $1-\beta$, que  a probabilidade de no rejeitarmos $H_0$ quando a hiptese alternativa $H_a$  verdadeira.

No  muito comum ver anlise de poder de teste, principalmente em modelos de regresso. Mas quando fazemos isso, **na prtica** estabelecemos $H_a$ como uma diferena *mnima* considerada *significativa* para o negcio. No confundir *significante* e *significativa*.

- **Significativa**  uma diferena que tem significado de negcio. Por exemplo, podemos dizer que uma diferena de alturas de 0,1 cm entre pessoas, por mais que possa ter *significncia* estatstica, no  *significativa* em termos da rea de conhecimento ou do negcio.
- **Significante**  uma diferena que aparece relevante em um teste de *significncia* estatstico. A significncia, no caso,  uma quantidade definida, que  o erro tipo I.

Note que  possvel achar diferenas *significantes* e *no significativas* bem como o contrrio.

A frmula *geral* da estatstica de teste  a seguinte:


$$ t = \frac{\bar{D}-\mu}{S_D\frac{1}{\sqrt{n}}} $$

E sob cada uma das alternativas, fica assim:

$H_0$: os desgastes so iguais - ou seja $\mu_0 = 0$

$$ t = \frac{\bar{D}-\mu_0}{S_D\frac{1}{\sqrt{n}}} = \frac{\bar{D}}{S_D\frac{1}{\sqrt{n}}}$$

$H_a$: A diferena entre os desgastes  *relevante*, igual a 10 - ou seja $\mu_a = 10$

$$ t = \frac{\bar{D}-\mu_a}{S_D\frac{1}{\sqrt{n}}} = \frac{\bar{D}-10}{S_D\frac{1}{\sqrt{n}}}$$


```python
fig, ax = plt.subplots(figsize=(10,3))

mu = 0
variance = 1
sigma = variance**.5

# Densidade da estatstica do teste sob H0
x0 = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
y0 = st.t.pdf(x, 18)
low = -np.inf
high = st.t.ppf(.05, n-1)

ax.plot(x0, y0)
ax.vlines(x=[t], 
             ymin=0, ymax=y.max(), 
             colors='r', linestyles='dotted')


plt.fill_between(x0, y0, where=(low < x0) & (x0 < high), alpha=.5)


# Densidade da estatstica do teste sob Ha
mua = -10/ep
xa = np.linspace(mua - 5*sigma, mua + 5*sigma, 100)
ya = st.t.pdf(xa-mua, 18)
lowa = high-mua
higha = np.inf

ax.plot(xa, ya)
ax.vlines(x=[t], 
             ymin=0, ymax=y.max(), 
             colors='r', linestyles='dotted')

plt.fill_between(xa, ya, where=(lowa < x0) & (x0 < higha), alpha=.5)

prob_area = st.t.cdf(high, n-1) - st.t.cdf(low, n-1)
print(f'A probabilidade para valores entre{low:.2f} e {high:.2f} '+\
      f' de {prob_area:.6%}'.replace('.',','))
```
