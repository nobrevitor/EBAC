```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 

warnings.filterwarnings('ignore')
sinasc_RO = pd.read_csv('b1abb33c4f.csv')
sinasc_RO.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.rc('figure', figsize = (6,4))
fig, axes = plt.subplots()
idade_mae = sns.boxplot(y= 'IDADEMAE',
            color = 'red',
            saturation = 0.4,
            data = sinasc_RO)
idade_mae.set_title('Boxplot Idade da Mãe')
idade_mae.set_ylabel('Idade da Mãe')
```

####  2) Faça um boxplot do peso do bebê

```python
plt.rc('figure', figsize = (6,4))
fig, axes = plt.subplots()
idade_mae = sns.boxplot(y= 'PESO',
            color = 'yellow',
            saturation = 0.4,
            data = sinasc_RO)
idade_mae.set_title('Boxplot Peso do Bebê')
idade_mae.set_ylabel('Peso (g)')
```

####  3) Faça um histograma da idade da mãe

```python
sinasc_RO[['IDADEMAE']].isnull().sum()
```

```python
sns.displot(x = 'IDADEMAE', 
            kde = True, 
            bins = 40,
            data = sinasc_RO)
plt.xlabel('Anos')
plt.ylabel('Quantidade')
plt.title('Idade da Mãe')
```
