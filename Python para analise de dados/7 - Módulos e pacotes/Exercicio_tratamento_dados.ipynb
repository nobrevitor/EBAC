<h3>Selecionando as linhas do dataframe original df em que a coluna stock é igual a MCD</h3>

```python
df_mcd = df[df['stock'] == 'MCD']
```

<h3>Selecionando apenas as colunas de data e valores de ações</h3>

```python
df_mcd = df_mcd[['date', 'open', 'high', 'low', 'close']]
```

```python
df_mcd.head(n=10)
```

Vamos limpar as colunas com o método apply, que permite a aplicação de uma função anônima (lambda) qualquer. A função lambda remove o caracter $ e faz a conversão do tipo de str para float.

```python
for col in ['open', 'high', 'low', 'close']:
  df_mcd[col] = df_mcd[col].apply(lambda value: float(value.split('$')[-1]))
```
