#### 1. Carregue os dados da tarefa anterior

````python
import yfinance as yf

import plotly.express as px
from plotly import graph_objects
from plotly.graph_objects import Layout

import pandas as pd
import numpy as np
````

````python
ativos = ['VALE3.SA', 'ITUB3.SA', 'PETR4.SA']

dt_ini = '2023-02-26'

df = yf.download(ativos, start=dt_ini)

df
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
df_close = df['Close']
df_close
````

````python
df_mean = df_close.rolling(90).mean()
df_mean_plot = df_mean.stack().reset_index()
df_mean_plot.columns = ['Data', 'Ativo', 'Valor']
````

````python
df_std = df_close.rolling(90).std()
df_std_plot = df_std.stack().reset_index()
df_std_plot.columns = ['Data', 'Ativo', 'Valor']
````

````python
fig = px.line(df_mean_plot, x='Data', y='Valor', color='Ativo')
fig.update_layout(title={
                'text': 'Gráfico médias móveis',
                'x': 0.5,
                'font': {'size': 20, 'color': 'black'}}, 
                  yaxis_title='Média do Valor')
fig.show()
````

````python
fig = px.line(df_std_plot, x='Data', y='Valor', color='Ativo')
fig.update_layout(title={
                'text': 'Gráfico Desvío Padrão Móvel dos Ativos',
                'x': 0.5,
                'font': {'size': 20, 'color': 'black'}}, 
                  yaxis_title='Desvío Padrão do Valor')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
indicadores = ['Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['PETR4.SA']*len(indicadores)
colunas = list(zip(indicadores, ativo))
````

````python
df_60 = df.iloc[-60:]
petr = df_60[colunas]
petr.columns = indicadores
petr
````

````python
grafico = graph_objects.Figure(
    data=[
        graph_objects.Candlestick(
            x=petr.index,
            open=petr['Open'],
            high=petr['High'],
            low=petr['Low'],
            close=petr['Close'],
            name='PETR4.SA'
        )
    ]
)

grafico.update_layout(
    title={
        'text': 'Índice Petrobrás',
        'x': 0.5,
        'font': {'size': 20, 'color': 'black'}},
    yaxis_title='Valor',
    paper_bgcolor='white',
    plot_bgcolor='white',
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='LightGrey'),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='LightGrey')
)
grafico.show()
````
